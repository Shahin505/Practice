```python
from sklearn.feature_selection import SelectKBest,f_classif
```

```python
score_culom=pd.DataFrame(fit_features.scores_,
columns=['score_value'])
```

```python
score_culom
```

```python
name_col=pd.DataFrame(x.columns)
```

```python
top_features=pd.concat([name_col,score_culom],axis=1)
```

```python
top_features
```

```python
top_features.nlargest(8,'score_value')
```
